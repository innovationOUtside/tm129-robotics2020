```python
import pandas as pd
```

```python
fn = 'typos.txt'
with open(fn,'r') as f:
    txt = f.readlines()
```

```python
txt
```

```python
# aspell
df = pd.DataFrame(columns=['filename', 'cell_type', 'typo'])

currfile = ''
cell_type = ''

for t in txt:
    t = t.strip('\n').strip()
    if not t or t in ['Misspelled words:', '!!!Spelling check failed!!!'] or t.startswith('-----'):
        continue
    
    if t.startswith('<htmlcontent>') or t.startswith('<py-'):
        if t.startswith('<html'):
            cell_type = 'md'
        elif t.startswith('<py-'):
            cell_type = 'code'
        else:
            cell_type=''

        currfile = t.split('/')[-1].split('.ipynb')[0]#+'.ipynb'
        continue
        
    df = df.append({'filename': currfile, 'cell_type': cell_type,
                    'typo': t}, ignore_index=True)
    
#df.drop_duplicates(inplace=True)
```

```python
laps_df = pd.DataFrame({'dt':['02:30.000', '02:10.000']})

# Cast to timedelta method 1
laps_df['_dt1'] = '0 days +00:'+laps_df['dt']
laps_df['_dt1'] = pd.to_timedelta(laps_df['_dt1'])

# Cast to timedelta method 3
to_s = lambda x: 60*int(x[0])+float(x[1])
laps_df['_dt2'] = laps_df['dt'].str.split(':').apply(to_s).astype(float)
laps_df['_dt2'] = pd.to_timedelta(laps_df['_dt2'], unit='s')

laps_df[['_dt1', '_dt2']].mean()
```

```python
def t_format(x):
    h, _r = divmod(x.total_seconds(), 3600)
    m, s = divmod(_r, 60)
    if h:
        return '{:02}:{:02}:{:02.3f}'.format(int(h), int(m), round(s,3))
    return '{:02}:{:02.3f}'.format(int(m), round(s,3))


def set_hms_format(x):
    # Make sure we have at least h:m:s
    x = '0:0:' + x
    # Timestring from last 3 fields
    _dt = ':'.join(x.split(':')[-3:])
    return pd.to_timedelta(_dt)
    
laps_df3 = pd.DataFrame({'dt':['30.1', '02:30.050', '01:02:10']})
laps_df3['_dt1'] = laps_df3['dt'].apply(set_hms_format)

laps_df3['_dt1'].apply(t_format)
```
